```python
# Change the name of column “ Quantity” to “Qty”
df.rename(columns={'Quantity' : 'Qty'},inplace=True)
df
```

```python
# Order the rows of transactions by Quantity ascending, TransactionDate descending
df.sort_values(by = ['Qty','TransactionDate'],ascending=[True,False])
```

```python
#  Set the column order of Transactions as ProductID, Quantity, TransactionDate, TransactionID, UserID
df[['PID','Qty','TransactionDate','TransactionID','UID']]
```

```python
#  Make UserID the first column of transactions
df[ ['UID'] + [  x for x in df.columns if x !='UID' ] ]
```

```python
# Extracting arrays from a Data Frame. Get the 2nd column
# list(df['PID'])
list(df.iloc[:,1:2])
list(df.iloc[:,1])

list(df.iloc[:,1])
```

```python
# Get the ProductID Array
list(df.columns).index('PID') # get the index of the PID
np.array(df.iloc[:,3])

df['PID'].tolist()
df['PID'].to_numpy()
```

```python
# Row subsetting, subset rows 1,3 and 6
df.iloc[[1,3,6]]

# using mask
foo = [False]*df.shape[0]
foo[1] = foo[3] = foo[6] = True
df[foo]
```

```python
# subset rows excluding 1,3 and 6
df.iloc[[x for x in df.index if x not in [1,3,6]]]
```

```python
# Subset the first three rows
df.iloc[[x for x in range(0,3)] ]
```

```python
# Subset rows excluding the last 2 rows
df.iloc[[x for x in range(-2,0,1)]]
```

```python
#  Subset rows excluding the first 3 rows
df.iloc[[x for x in range(3,len(df))]]
```

```python
#  Subset rows where Quantity>1
df[df['Qty'] > 1]
```

```python
# Subset rows where UserID=2
df[df['UID'] == 2]
```

```python
#  Subset rows where Quantity>1 and UserID=2
df[(df['Qty'] > 1) & (df['UID'] == 2)]
```

```python
#  Subset rows where Quantity + UserID is > 4
df[df['Qty']+df['UID'] > 4]
```

```python
# Subset rows where an external array, foo, is True
foo = [ True if(x%2) else False for x in range(len(df))]
df[foo]

foo = [x%2 == 0 for x in range(len(df))]
df[foo]
```

```python
# Subset rows where an external array, bar, is positive
bar = [1, -2, 3, -4, 5,0,2,3,1,3]
df[np.array(bar) > 0]
```

```python
# Subset rows where foo is TRUE or bar is negative
foo = np.array([True]*len(df))
bar = np.array(bar)
mask  = ( foo == True) | (bar < 0)
df.iloc[mask]

#in single line
df.iloc[ ( np.array([True]*len(df)) == True | np.array([1]*len(df)) ) ]
```

```python
# Subset the rows where foo is not TRUE and bar is not negative
df.iloc[  ((~foo) | (bar >=  0) )  ]
```

```python
# Subset by columns 1 and 3
df.iloc[:,[1,3]]
```

```python
# Subset by columns TransactionID and TransactionDate
df[['TransactionID','TransactionDate']]
```

```python
# Subset by columns TransactionID and TransactionDate with logical operator
df.iloc[:,(df.columns == 'TransactionID') | (df.columns == 'TransactionDate')] # we get subset mask
```

```python
# Subset columns by a variable list of column names
list = ['TransactionID','TransactionDate']
df[list]
```

```python
# Subset columns excluding a variable list of column names
list = ['TransactionID','TransactionDate']
df[[x for x in df.columns if x not in list]]
```

```python
# Inserting and updating values

# inserting
new_row = [1,'0000-00-00',7.2,2,1000]
df.loc[len(df)] = new_row
df

# updating

df.loc[df['Qty'] == 1000 ,'TransactionID']  = 200
df
```
